```python
# Dependencies
import json
import requests
import pprint
import pandas as pd
from config import api_key
```

```python
# Save config information
url = "https://api.covidactnow.org/v2/states.json?"

# Build query URL
query_url = url + "apiKey=" + api_key 
```

```python
# Get covid data
state_response = requests.get(query_url)
state_json = state_response.json()

print(f"{state_json}")
```

```python
df = pd.DataFrame(state_json)
df.head(5)
```

```python
df.set_index(['fips'])
```

```python
filtered_df = df.loc[:, ['state', 'county', 'metrics', 'actuals']]
filtered_df.head()
```

```python
metrics = filtered_df['metrics']
```

```python
metrics = metrics.to_json()
```

```python
metrics_dict = json.loads(metrics)
```

```python
metrics_dict
```

```python
metrics_df = pd.DataFrame.from_dict(metrics_dict, orient='index')
metrics_df.head()
```
