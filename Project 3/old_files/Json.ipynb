```python
# Dependencies
import json
import requests
import pprint
import pandas as pd
from config import api_key
```

```python
# Save config information
url = "https://api.covidactnow.org/v2/states.json?"

# Build query URL
query_url = url + "apiKey=" + api_key 
```

```python
# Get weather data
state_response = requests.get(query_url)
state_json = state_response.json()

print(f"{state_json}")
```

```python
df = pd.DataFrame(state_json)
df
```
